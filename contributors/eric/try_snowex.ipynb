```python
# Import our api class
import xarray as xr
from snowexsql.api import LayerMeasurements
from datetime import datetime
import pandas as pd
import geopandas as gpd 
import matplotlib.pyplot as plt

# import some gis functionality 
from shapely.geometry import Point 
```

```python
measurements = LayerMeasurements()

# Get the unique data names/types in the table
results = measurements.all_types
print('Available types = {}'.format(', '.join([str(r) for r in results])))

# Get the unique instrument in the table
results = measurements.all_instruments
print('\nAvailable Instruments = {}'.format(', '.join([str(r) for r in results])))

# Get the unique dates in the table
results = measurements.all_dates
print('\nAvailable Dates = {}'.format(', '.join([str(r) for r in results])))

# Get the unique site names in the table
results = measurements.all_site_names
print('\nAvailable sites = {}'.format(', '.join([str(r) for r in results])))
```

```python
df = LayerMeasurements.from_filter(site_name='Grand Mesa',type='lwc_vol',limit=3000)
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.longitude,df.latitude),crs=4326)
gdf
```

```python
gdf.explore(tiles='Esri WorldImagery')
```

```python
# convert to gdf
```

```python
gdf['date'] = pd.to_datetime(gdf['date'])
```

```python
gdf = gdf.set_index(['site_id','date','depth'])
gdf = gdf[~gdf.index.duplicated()]
#df = df.set_index(['pit_id','depth'])
gdf
```

```python
# Convert the DataFrame to an xarray Dataset
ds = xr.Dataset.from_dataframe(gdf)
ds
```

```python
ds['site_id']=ds['site_id'].astype(str)
ds['value'] = ds['value'].astype(float)
```

```python
da = ds['value'].isel(site_id=slice(0,20))
da
```

```python
# Create a figure with subplots for each site
fig, axes = plt.subplots(nrows=da.sizes['site_id'], ncols=1, figsize=(15, 5 * da.sizes['site_id']), sharex=True, sharey=True)

# Loop through each site and create a plot
for i, site_id in enumerate(da.site_id):
    ax = axes[i]
    site_data = da.sel(site_id=site_id)
    site_data.plot.imshow(ax=ax, x='date', y='depth', cmap='viridis', add_colorbar=(i == 0))
    ax.set_title(f'Site ID: {site_id.values}')
    ax.set_xlabel('Date')
    ax.set_ylabel('Depth')

plt.tight_layout()
plt.show()
```
